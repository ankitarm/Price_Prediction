```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
mpl.rcParams["figure.figsize"] = (20,10)
df9 = pd.read_csv("BHD_NoOutliers.csv")
df9.head()
```

```python
#dropping unneccessary columns
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head(3)
```

```python
#location cannot interprete text column, so we convert to numerical 

dummies = pd.get_dummies(df10.location)
dummies.head()
```

```python
#now concating the two dataframes
#to avoid dummy trap dropping last column(we can keep all dummies value 0 to represent last column)
df11 = pd.concat([df10,dummies.drop('other',axis='columns')],axis='columns')
df11.head()
```

```python
#now dropping location column
df12 = df11.drop('location',axis='columns')
df12.head()
```

```python
df12.shape
```

```python
#consider the dependent columns drop rest
X = df12.drop(['Unnamed: 0','price'], axis='columns')
X.head()
```

```python
#price is dependent because it depends on the area
y = df12.price
y.head()
```

```python
#splitting the arrary into subarrays for testing and training, here 20% is for testing and 80% for training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```
